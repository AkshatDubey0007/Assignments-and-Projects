<h1><font color=Brown>3. Replacing the null values by mean of the column. </font></h1>

```python
df1.head()
summary_mean=df1["summary.score"].mean()
print(summary_mean)
df1["summary.score"].replace(np.nan,summary_mean)
df1.head()
```

```python
rating_mean="Very Good"
print(rating_mean)
df1["rating.band"]=df1["rating.band"].replace(np.nan,rating_mean)
df1.head()
```

```python
df1.head()
atmosphere_mean=df1["atmosphere"].mean()
print(atmosphere_mean)
df1["atmosphere"]=df1["atmosphere"].replace(np.nan,atmosphere_mean)
df1.head()
```

```python
df1.head()
cleanliness_mean=df1["cleanliness"].mean()
print(cleanliness_mean)
df1["cleanliness"]=df1["cleanliness"].replace(np.nan,cleanliness_mean)
df1.head()
```

```python
df1.head()
facilities_mean=df1["facilities"].mean()
print(facilities_mean)
df1["facilities"]=df1["facilities"].replace(np.nan,facilities_mean)
df1.head()
```

```python
df1.head()
security_mean=df1["security"].mean()
print(security_mean)
df1["security"]=df1["security"].replace(np.nan,security_mean)
df1.head()
```

```python
df1.head()
staff_mean=df1["staff"].mean()
print(staff_mean)
df1["staff"]=df1["staff"].replace(np.nan,staff_mean)
df1.head()
```

<h1><font color=Brown>4. Having a quick view of the replaced values</font></h1>

```python
df1.head()
value_mean=df1["valueformoney"].mean()
print(value_mean)
df1["valueformoney"]=df1["valueformoney"].replace(np.nan,value_mean)
df1.head(50)
```
